# Task_5

Напишите функцию, которая формирует два массива $x$ и $y$ для построение окружности заданного радиуса и центром в начале координат.

Для этого угдобно пользоваться полярными координатами. Уравнение окружности в полярных координатах будет выглядеть

$$\rho = R = const$$

$$\rho \neq f(\varphi)$$

$x$ и $y$ будут проекциями радиус-вектора $R$ для всех $\varphi \in [0;\;\;2\pi]$

Постройте эту окружность!

```python
def circle(r:float)-> tuple:
  t = np.linspace(0, 2*np.pi, 100)
  x = r*np.cos(t)
  y = r*np.sin(t)
  return x,y
```

```python
plt.plot(x,y)
```

Напишите программу, которая строит такую спираль:

image.png

```python
def nine(a:float, b:float, c: float) -> tuple:
  t = np.linspace(0, 2*np.pi, 100)
  x = a * np.cos(t)
  y = b * np.sin(t) - c
  return x,y
```

```python
nine = nine(2,1,0.5)
```

```python
plt.plot(nine[0], nine[1])
```

```python
# Параметры архимедовой спирали
a = 0.1  # коэффициент

# Параметрические уравнения архимедовой спирали
t = np.linspace(0, np.pi, 1000)
x = a * t * np.cos(t)
y = a * t * np.sin(t)

# Построение траектории спирали
plt.plot(x, y, label='Архимедова спираль')

# Добавление остальных деталей графика
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('Траектория архимедовой спирали')
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```
