```python
def arr(n_row: int =300)->np.ndarray:
  return np.random.randn(n_row, 2)
matrix = np.concatenate([arr(), arr() + np.array([4, -5])], axis=0)
```

```python
def nine(a:float, b:float, c: float) -> tuple:
  t = np.linspace(0, 2*np.pi, 100)
  x = a * np.cos(t)
  y = b * np.sin(t) - c
  return x,y
```

```python
nine = nine(2,1,0.5)
```
